```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_data=pd.read_csv(r"C:\Users\The Doctor\Anaconda3\envs\PythonData\mouse_drug_data.csv")
mouse_data
```

```python

clinical_data=pd.read_csv(r"C:\Users\The Doctor\Anaconda3\envs\PythonData\clinicaltrial_data.csv")
clinical_data
```

```python

# Merge two dataframes 
merge_mouse = pd.merge(mouse_data, clinical_data, on="Mouse ID")
merge_mouse
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_avg_tumor_vol = merge_mouse[["Drug", "Timepoint","Tumor_Volume"]]
grouped_drug_avg_tumor_vol = drug_avg_tumor_vol.groupby(['Drug', 'Timepoint'])
```

```python
# Convert to DataFrame
tumor_vol = pd.DataFrame(grouped_drug_avg_tumor_vol["Tumor_Volume"].agg(np.mean))
tumor_vol.head()
```

```python



# Preview DataFrame
tumor_vol_indexed = tumor_vol.reset_index()
tumor_vol_indexed
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_vol2 = merge_mouse.groupby(['Drug', 'Timepoint']).sem()
tumor_vol2
```

```python

# Convert to DataFrame
tumor_vol2_pd = pd.DataFrame(tumor_vol2[["Tumor_Volume"]])

# Preview DataFrame
tumorvol_sem = tumor_vol2_pd.reset_index()
tumorvol_sem.head()
```
